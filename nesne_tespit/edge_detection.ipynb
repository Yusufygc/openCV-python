```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
img = cv2.imread('C:\\Users\\TUF Dash F15\\Desktop\\cvCourse\\images\\dog_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127) # threshold1 and threshold2 are the min and max values
```

```python
plt.imshow(edges) # ok iyi olmad grntde ok grlt var
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255) # threshold deerlerini deitirerek grlty azaltabiliriz ama en salkls blur uygulamak
```

```python
blurred_img = cv2.blur(img, ksize=(5,5)) # ksize = kernel size
```
