```python
face_cascade = cv2.CascadeClassifier('C:\\Users\\TUF Dash F15\\Desktop\\cvCourse\\haarcascades\\haarcascade_frontalface_default.xml')
```

```python
def detect_face(img): # resimdeki yzleri bulan fonksiyon
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5) # scaleFactor: resmin her seferinde kltlme oran, minNeighbors: her yz iin ka komu bulunmal (yzn etrafnda ka tane kare bulunmal) onu belirler
    for (x,y,w,h) in face_rects: # yzn koordinatlar
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,0),10)
    return face_img
```

```python
result = detect_face(erkek)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result1 = detect_face(kadn)
```

```python
plt.imshow(result1,cmap='gray')
```

```python
result2 = detect_face(toplu_foto)
```

```python
plt.imshow(result2,cmap='gray') 
```

```python
eye_cascade=cv2.CascadeClassifier('C:\\Users\\TUF Dash F15\\Desktop\\cvCourse\\haarcascades\\haarcascade_eye.xml')
```

```python
def detect_eye(img): # resimdeki yzleri bulan fonksiyon
    face_img = img.copy()
    eyes_rects = eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5) # scaleFactor: resmin her seferinde kltlme oran, minNeighbors: her yz iin ka komu bulunmal
    for (x,y,w,h) in eyes_rects: # yzn koordinatlar
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,0),10)
    return face_img
```

```python
result=detect_eye(erkek)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result=detect_eye(kadn)
plt.imshow(result,cmap='gray')
```

```python
result=detect_eye(toplu_foto)
plt.imshow(result,cmap='gray')
```
