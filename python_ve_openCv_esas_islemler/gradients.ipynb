```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('C:\\Users\\TUF Dash F15\\Desktop\\cvCourse\\images\\sudokuu.jpg',0)
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) # x sobel iin 1  verdik y 0 oldu 
# Sobel X Gradient (detects vertical edges) 
#  64F is the data type of the output image (64 bit float) 
#  1 is the x direction 
#  0 is the y direction 
#  ksize is the kernel size (5x5) 
#  Sobel is a combination of Gaussian smoothing and differentiation 
#  Sobel is more resistant to noise 
#  Sobel is a 3x3 kernel 
#  Sobel is a 2nd derivative 
#  Sobel is a 2nd order derivative 
#  Sobel is a 2nd order derivative in both x and y direction
#  Sobel is a 2nd order derivative in both x and y direction and then we combine them 
#  Sobel is a 2nd order derivative in both x and y direction and then we combine them to get the final result

# Sobel X Gradient (dikey kenarlar alglar)
# 64F, kt grntsnn veri trdr (64 bit kayan nokta) duyarll artrr
# 1 x yndr
# 0 y yndr
# ksize ekirdek boyutudur (5x5)
# Sobel, Gauss yumuatma ve farkllamann birleimidir
# Sobel grltye kar daha dayankldr
# Sobel 3x3'lk bir ekirdektir
# Sobel 2. trevdir
# Sobel 2. dereceden bir trevdir
# Sobel hem x hem de y ynnde 2. dereceden trevdir
# Sobel hem x hem de y ynnde 2. dereceden trevdir ve bunlar birletiriyoruz
# Sobel hem x hem de y ynnde 2. dereceden bir trevdir ve sonra bunlar birletirip nihai sonucu elde ederiz
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_img(sobely)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
blended=cv2.addWeighted(src1=sobelx, alpha=0.5,src2=sobely,beta=0.5,gamma=0) # x ve y ynnde 2. dereceden bir trevdir ve sonra bunlar birletirip nihai sonucu elde ederiz
```

```python
display_img(blended)
```

```python
ret ,th1 = cv2.threshold(img,40,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)# 4x4'lk bir ekirdek oluturduk
```

```python
gradient =cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel) # ama ilemi rakamlarnn iini beyaz ile doldurduk
```

```python
display_img(gradient)
```
