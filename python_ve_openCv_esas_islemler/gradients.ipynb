```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('C:\\Users\\TUF Dash F15\\Desktop\\cvCourse\\images\\sudokuu.jpg',0)
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) # x sobel için 1  verdik y 0 oldu 
# Sobel X Gradient (detects vertical edges) 
#  64F is the data type of the output image (64 bit float) 
#  1 is the x direction 
#  0 is the y direction 
#  ksize is the kernel size (5x5) 
#  Sobel is a combination of Gaussian smoothing and differentiation 
#  Sobel is more resistant to noise 
#  Sobel is a 3x3 kernel 
#  Sobel is a 2nd derivative 
#  Sobel is a 2nd order derivative 
#  Sobel is a 2nd order derivative in both x and y direction
#  Sobel is a 2nd order derivative in both x and y direction and then we combine them 
#  Sobel is a 2nd order derivative in both x and y direction and then we combine them to get the final result

# Sobel X Gradient (dikey kenarları algılar)
# 64F, çıktı görüntüsünün veri türüdür (64 bit kayan nokta) duyarlılığı artırır
# 1 x yönüdür
# 0 y yönüdür
# ksize çekirdek boyutudur (5x5)
# Sobel, Gauss yumuşatma ve farklılaşmanın birleşimidir
# Sobel gürültüye karşı daha dayanıklıdır
# Sobel 3x3'lük bir çekirdektir
# Sobel 2. türevdir
# Sobel 2. dereceden bir türevdir
# Sobel hem x hem de y yönünde 2. dereceden türevdir
# Sobel hem x hem de y yönünde 2. dereceden türevdir ve bunları birleştiriyoruz
# Sobel hem x hem de y yönünde 2. dereceden bir türevdir ve sonra bunları birleştirip nihai sonucu elde ederiz
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_img(sobely)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
blended=cv2.addWeighted(src1=sobelx, alpha=0.5,src2=sobely,beta=0.5,gamma=0) # x ve y yönünde 2. dereceden bir türevdir ve sonra bunları birleştirip nihai sonucu elde ederiz
```

```python
display_img(blended)
```

```python
ret ,th1 = cv2.threshold(img,40,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)# 4x4'lük bir çekirdek oluşturduk
```

```python
gradient =cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel) # açma işlemi rakamlarının içini beyaz ile doldurduk
```

```python
display_img(gradient)
```
