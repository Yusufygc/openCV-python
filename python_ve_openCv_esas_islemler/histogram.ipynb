```python
img= bricks # resmi img deikenine atadk
```

```python
color = ('b', 'g', 'r') # renklerin histogram iin liste oluturduk 

for i, col in enumerate(color): # enumerate = index ve deer dndrr (0, 'b') (1, 'g') (2, 'r') gibi
    hist = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist, color=col) # renklerin histogramn izdirdik
    plt.xlim([0,256]) # x eksenini 0-256 arasnda snrladk

plt.title('histogram') # balk
```
