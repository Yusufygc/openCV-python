```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
bos_foto = np.zeros((512,512,3), dtype=np.int16) # 512x512 boyutunda siyah bir resim oluşturur. np.int16 ile 16 bitlik bir resim oluştururuz.
```

```python
plt.imshow(bos_foto) # resmi gösterir.
```

```python
cv2.rectangle(bos_foto, pt1=(300,100), pt2=(400,300), color=(255,0,0), thickness=7) # dikdörtgen çizer. pt1 ve pt2 ile başlangıç ve bitiş noktalarını belirleriz. color ile renk belirleriz. thickness ile kalınlık belirleriz.
```

```python
plt.imshow(bos_foto) # resmi gösterir.
```

```python
cv2.rectangle(bos_foto, pt1=(200,200), pt2=(300,300), color=(0,255,0), thickness=10) # dikdörtgen çizer. pt1 ve pt2 ile başlangıç ve bitiş noktalarını belirleriz. color ile renk belirleriz. thickness ile kalınlık belirleriz.
plt.imshow(bos_foto) # resmi gösterir.
```

```python
cv2.circle(bos_foto, center=(100,100), radius=50, color=(0,0,255), thickness=8) # daire çizer. center ile merkez noktasını belirleriz. radius ile yarıçapı belirleriz. color ile renk belirleriz. thickness ile kalınlık belirleriz.
plt.imshow(bos_foto) # resmi gösterir.
```

```python
cv2.circle(bos_foto, center=(100,100), radius=50, color=(0,0,255), thickness=-8) # daire çizer. center ile merkez noktasını belirleriz. radius ile yarıçapı belirleriz. color ile renk belirleriz. thickness ile kalınlık belirleriz. kalınlığı eksi değer verirsek içini doldurur.
plt.imshow(bos_foto) # resmi gösterir.
```

```python
cv2.line(bos_foto, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5) # çizgi çizer. pt1 ve pt2 ile başlangıç ve bitiş noktalarını belirleriz. color ile renk belirleriz. thickness ile kalınlık belirleriz.
plt.imshow(bos_foto) # resmi gösterir.
```
