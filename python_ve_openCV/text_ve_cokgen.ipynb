```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
bos_foto =np.zeros((512,512,3),dtype=np.int16)
plt.imshow(bos_foto)
```

```python
cv2.putText(bos_foto,text="Merhaba",org=(10,500),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA) #org = orgin yani başlangıç noktası  #fontScale = yazı boyutu #thickness = yazı kalınlığı #lineType = yazı tipi  #cv2.LINE_AA = yazı tipi düzgün olsun diye yazıyoruz  #color = yazı rengi #fontFace = yazı tipi
plt.imshow(bos_foto)
```

```python
noktalar =np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
```

```python
noktalar.shape
```

```python
cv2.polylines(bos_foto,[noktalar],isClosed=True,color=(255,255,255),thickness=5) #isClosed = kapalı mı değil mi #color = renk #thickness = kalınlık #noktaları birleştirip çizgi çiziyor 
plt.imshow(bos_foto)
```

```python
noktalar =np.array([[300,150],[50,50],[500,200],[20,40]],np.int32)
cv2.polylines(bos_foto,[noktalar],isClosed=True,color=(255,255,255),thickness=5) #isClosed = kapalı mı değil mi #color = renk #thickness = kalınlık #noktaları birleştirip çizgi çiziyor 
plt.imshow(bos_foto)
```
